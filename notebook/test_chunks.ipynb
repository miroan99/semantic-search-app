```python
from pathlib import Path

# import your function
from app.utils import chunk_text

# test cases
samples = {
    "short": "hej med dig",
    "medium": "Dette er en lidt længere tekst der skal chunkes i bidder. "
              "Formålet er at se, om funktionen stopper korrekt, også når teksten er kortere end overlap.",
    "long": "A" * 2500  # simulate long text
}

for name, txt in samples.items():
    chunks = chunk_text(txt, max_chars=800, overlap=100)
    print(f"\n{name.upper()} -> {len(chunks)} chunks")
    for i, ch in enumerate(chunks):
        print(f"  chunk {i}: {len(ch)} chars")
```
